```python
carros = pd.read_csv('cars.csv', usecols=['Brand','Body.Type', 'Power.hp','Height.in', 'Torque.lbft'])
print(carros)
```

```python
print(carros[carros['Body.Type'] == 'Hatchback'])
```

```python
print(carros['Body.Type'].unique())
```

```python
print(len(carros['Body.Type'].unique()))
```

```python
carros.dropna(subset=['Power.hp','Height.in', 'Torque.lbft'], inplace=True)
print(carros.info())
```

```python
tipo_auto=['Coupe', 'SUV', 'Van', 'Hatchback']
carros = carros[carros['Body.Type'].isin(tipo_auto)]
cars2= carros.iloc[0:2601]
print(carros.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,8))
sns.set_style(rc={'axes.facecolor':'EBEBEB'})
sns.scatterplot(x='Torque.lbft', y='Height.in', data=cars2)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.set_style(rc={'axes.facecolor':'EBEBEB'})
sns.scatterplot(x='Torque.lbft', y='Height.in', data=cars2, s=100, hue='Body.Type')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Modelos
modeloknn= KNeighborsClassifier(n_neighbors=3)
x=cars2[['Power.hp','Height.in']]
y=cars2['Body.Type']
modeloknn.fit(x,y)
```

```python
#Copue
auto1 = {'Power.hp':[296], 'Height.in':[51.6]}
auto1df=pd.DataFrame(auto1)
casif1=modeloknn.predict(auto1df)
print(casif1)
```

```python
#suv
auto2 = {'Power.hp':[280], 'Height.in':[66]}
auto2df=pd.DataFrame(auto2)
casif2=modeloknn.predict(auto2df)
print(casif2)
```

```python
#suv
auto3 = {'Power.hp':[110], 'Height.in':[60]}
auto3df=pd.DataFrame(auto3)
casif3=modeloknn.predict(auto3df)
print(casif3)
```

```python
plt.figure(figsize=(10,8))
sns.set_style(rc={'axes.facecolor':'EBEBEB'})
sns.scatterplot(x='Power.hp', y='Height.in', data=cars2, s=100, hue='Body.Type')
sns.scatterplot(x=auto1df['Power.hp'], y=auto1df['Height.in'], color='red', s=200)
sns.scatterplot(x=auto2df['Power.hp'], y=auto2df['Height.in'], color='blue', s=200)
sns.scatterplot(x=auto3df['Power.hp'], y=auto3df['Height.in'], color='black', s=200)
plt.show()
```
