```python
import pandas as pd
```

```python
#Lectura de datos
precios = pd.read_csv("C:\\Users\\cldl\\Downloads\\OptaIII\\precios.csv")
print(precios.Year)
print(" ")
print(precios.Australia)
```

```python
#Escritura
datos = {'Nombre': ['Maria', 'Pedro', 'Alexandra','Juan','Ximena','Xavier'],
         'Apellido': ['Castro', 'Aro', 'Bueno', 'lopez', 'Frias', 'Osorio'],
         'Edad': [19, 24, 20, 21, 25, 22],
        'Salario': [800.50,900.00,705.25,1025.38,987.65,1250.45]}
df = pd.DataFrame(datos, columns=['Nombre', 'Apellido', 'Edad', 'Salario'])
df.to_csv("C:\\Users\\cldl\\Downloads\\OptaIII\\nomina.csv")
```

```python
#Grfica de precios
plt.figure(figsize=(10, 5))
plt.plot(precios.Year, precios.Australia, 'r', linestyle='dashed', label='Australia')
plt.plot(precios.Year, precios.Canada, 'b', label='Canada')
plt.plot(precios.Year, precios.France, '#00CCFF', label='France')
plt.title('Precios del Arroz')
plt.xlabel('Ao')
plt.ylabel('Precio')
plt.legend()
plt.show()
```
