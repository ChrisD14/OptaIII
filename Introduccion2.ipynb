```python
import pandas as pd
```

```python
signos = pd.read_excel('C:\\Users\\cldl\\Downloads\\OptaIII\\signosvitales.xlsx')
print(signos)
```

```python
new_signos = signos.dropna()#Borra celdas vacias
print(new_signos)
```

```python
signosna = signos.copy()#Copia el dataframe original
signosna.fillna(100, inplace=True)#Rellena celdas vacias
print(signosna)
```

```python
signosnacal = signos.copy()
# Fill missing values in the 'Calorias' column with 50
signosnacal.fillna({'Calorias': 50}, inplace=True)
print(signosnacal)
```

```python
signosp = signos.copy()
signosp.loc[1, 'Calorias'] = 1000
print(signosp)
```

```python
#Seleccion de Columnas
print(signos.Duracion)
```

```python
print(signos[["Duracion","Fecha"]])
```

```python
#Oredenar los datos
signos.sort_values(by='Duracion', ascending=True, inplace=True)
print(signos)
```

```python
signos.sort_values(by='Pulso', ascending=True, inplace=True)
print(signos)
```

```python
signos.sort_values(by='Pulso', ascending=False, inplace=True)
print(signos)
```

```python
signos.drop_duplicates()
print(signos)
```

```python
signos.drop_duplicates(inplace=True)
print(signos)
```

```python
media=signos['Calorias'].mean()
print("Media Calorias: ",media)
```

```python
signos['Calorias'].fillna(media, inplace=True)
print(signos)
```
