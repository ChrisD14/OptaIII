```python
h=[5, 7, 9, 10, 11, 14, 13, 15, 17, 17, 18, 16, 19, 15, 17]
y=[82, 84, 86, 88, 88, 90, 89, 94, 95, 93, 98, 97, 95, 89, 88]
plt.figure(figsize=(10, 5))
plt.scatter(h,y, color='red')
##plt.plot(h,y)
plt.show()
```

```python
slope, int, a, p,std=stats.linregress(h,y)
def funcion(x):
    return slope*x + int
modelo=list(map(funcion, h))
plt.figure(figsize=(10, 5))
plt.scatter(h,y, color='red')
plt.plot(h,modelo, color='blue')
plt.show()
```

```python
h=[1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y=[100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99 , 99, 100]
modelo=np.poly1d(np.polyfit(h,y,3))
linea=np.linspace(1, 22, 100)
plt.figure(figsize=(10, 5))
plt.scatter(h,y, color='red')
plt.plot(linea,modelo(linea))
plt.show()
```
