```python
import pandas as pd
```

```python
venta = pd.read_csv('C:\\Users\\cldl\\Downloads\\OptaIII\\ventas.csv')
print(venta)
```

```python
venta.info()
```

```python
lista_columna = list(venta.columns)
print(lista_columna)
```

```python
#Eliminar la columna 'Season'
venta.drop(columns=['Season'], inplace=True)
print(venta)
```

```python
venta.info()
```

```python
print(venta.isna().sum())
```

```python
ventas= venta.dropna(subset=["Season"])
print(ventas.isna().sum())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.pie(ventas['Gender'].value_counts(), labels=venta['Gender'].value_counts().index,
        autopct='%1.2f%%', radius =1)
plt.title('Ventas H vs M')
plt.legend()
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(6, 4))
plt.pie(ventas['Gender'].value_counts(), labels=venta['Gender'].value_counts().index,
        autopct='%1.2f%%', radius =1)
plt.title('Ventas H vs M')
plt.legend()
plt.show()
```

```python
#Grafica de productos
producto = venta['Items Purchased'].value_counts()
plt.figure(figsize=(10, 6))
producto.plot(kind='bar', color='blue')
plt.title('Productos vendidos')
plt.show()
```

```python
#Grafica de productos
producto = venta['Items Purchased'].value_counts()
plt.figure(figsize=(10, 6))
producto.plot(kind='bar', color='skyblue')
plt.title('Distribucin de productos')
plt.xlabel('Productos')
plt.ylabel('Cantidad de ventas')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#Disribucion de las edades
venta['Grupo Edad'] = pd.cut(venta['Age'], bins=[0, 18, 25, 35, 45, 55, 65, float('inf')], labels=['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']
                             , include_lowest=True)
edad=venta.pivot_table(index='Grupo Edad', columns='Gender', aggfunc='size', fill_value=0)
edadgx=edad.plot(kind='bar')
plt.show()
```
